**15: Add a title to each of the two plots youâ€™ve created, and adjust the margins to make the text youâ€™ve added look better.**

````python
plt.figure(figsize=(12, 8))

ax1 = plt.subplot(1, 2, 1)
x_values = range(len(months))
plt.plot(x_values, visits_per_month, marker='s')
plt.xlabel('Months')
plt.ylabel('Visits')
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)

# code for task15(title for the subplot left)
plt.title('Total Visits per Month')

ax2 = plt.subplot(1, 2, 2)
plt.plot(x_values, key_limes_per_month, color='red', marker='o')
plt.plot(x_values, persian_limes_per_month, color='blue', marker='o')
plt.plot(x_values, blood_limes_per_month, color='green', marker='o')
plt.xlabel('Months')
plt.ylabel('Numbers of limes sold')
plt.legend(['Key Limes', 'Persian Limes', 'Blood Limes'])
ax2.set_xticks(x_values)
ax2.set_xticklabels(months)

# code for task15(title for the subplot right, and subplots adjustment)
plt.title('Number of different limes sold each month')
plt.subplots_adjust(wspace=0.3)

plt.show()
````

**16: Now, save your figure as a png with a descriptive file name.**

````python
plt.figure(figsize=(12, 8))

ax1 = plt.subplot(1, 2, 1)
x_values = range(len(months))
plt.plot(x_values, visits_per_month, marker='s')
plt.xlabel('Months')
plt.ylabel('Visits')
ax1.set_xticks(x_values)
ax1.set_xticklabels(months)
plt.title('Total Visits per Month')

ax2 = plt.subplot(1, 2, 2)
plt.plot(x_values, key_limes_per_month, color='red', marker='o')
plt.plot(x_values, persian_limes_per_month, color='blue', marker='o')
plt.plot(x_values, blood_limes_per_month, color='green', marker='o')
plt.xlabel('Months')
plt.ylabel('Numbers of limes sold')
plt.legend(['Key Limes', 'Persian Limes', 'Blood Limes'])
ax2.set_xticks(x_values)
ax2.set_xticklabels(months)
plt.title('Number of different limes sold each month')

plt.subplots_adjust(wspace=0.3)

# code for task16
plt.savefig('number_of_visitors_and_limes.png')

plt.show()
````
